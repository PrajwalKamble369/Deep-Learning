```python
"""
Task 1

Create tensors using:

empty
zeros
ones
rand
tensor
arange
linspace

without looking at notes.
"""
import torch

empty_tensor = torch.empty((0))
print("empty tensor",empty_tensor,"checking datatype", type(empty_tensor))

zeros_tensor = torch.zeros(4,4)
print("zeros tensor\n",zeros_tensor)

ones_tensor = torch.ones((3,3))
print("ones tensor \n", ones_tensor)

rand_tensor = torch.rand(6,2)
print("rand tensor \n",rand_tensor)

arange_tensor = torch.arange(5)
print("arange tensor\n", arange_tensor)

linspace_tensor = torch.linspace(-20,20,5)
print("linspace tensor\n", linspace_tensor)
```
