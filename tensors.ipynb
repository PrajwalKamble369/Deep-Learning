# <u>**Tensors**</u>

- What are Tensors?
    - Tensor is data strcture, it is specialized multi-dimensional array designed for mathematical and computational efficiency.
    - Dimesion in tensor is, how span of tensor in directions.

- Real World Examples of Tensors:
    1) Scaler : 0 Dimensional Tensor
        - Single Number
    2) Vector : 1 Dimensional Tensor
        - Sequence or collection (list of number)
    3) Metrices : 2 Dimensional Tensor
        - Represent Tabular or grid type data
    4) 3D Tensor : Coloured Image
        - RGB Image
    5) 4D Tensor : Batches of RGD image
        - a 
    6) 5D Tensor : Video Data
        - b
